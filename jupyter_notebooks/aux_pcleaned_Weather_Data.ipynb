```python
import pandas as pd
```

```python
weather_data = pd.read_csv('../data/raw/GlobalWeatherRepository.csv')
weather_data.head()
```

```python
weather_data = weather_data[['country', 'temperature_celsius','last_updated']]
weather_data['last_updated'] = pd.to_datetime(weather_data['last_updated'])
weather_data['W_Date'] = weather_data['last_updated'].dt.date
weather_data['W_Date'] = pd.to_datetime(weather_data['W_Date'])
weather_data.drop(columns=['last_updated'], inplace=True)
```

```python
weather_data['Temp_month'] = weather_data['W_Date'].dt.month
weather_data['Temp_year'] = weather_data['W_Date'].dt.year
```

```python
weather_data.head()
```

```python
weather_data.to_csv('../data/processed/Temperature Data.csv', index=False)
```
